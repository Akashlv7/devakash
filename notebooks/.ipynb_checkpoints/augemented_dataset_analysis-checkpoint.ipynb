```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
tv_data = pd.read_csv("tv-audience-dataset.csv")
```

```python
print (tv_data.shape)
print (tv_data.head())
```

```python
tv_data.columns = ["channel_id", "slot", "week", "genre_id", "sub_genre_id", "user_id", "program_id", "event_id", "duration"]
```

```python
#Weeks 14 and 19 should not be used because they contain errors.
corr_tv_data = tv_data[(tv_data["week" ] <14) & (tv_data["duration" ] > 10 ) & (tv_data["genre_id" ] != 7 ) & (tv_data["sub_genre_id" ] != 47 ) ]
```

```python
corr_tv_data["hour"] = corr_tv_data["slot"] % 24
corr_tv_data =corr_tv_data.replace({'hour': 0}, 24)

corr_tv_data["day"] = (abs(corr_tv_data["slot"]-1) / 24).astype(int) + 1
```

```python
sns.distplot(corr_tv_data["day"]);
print (corr_tv_data["day"].value_counts())
```

```python
plt.figure(figsize=(36, 16))
sns.distplot(corr_tv_data["hour"]);
print (corr_tv_data["hour"].value_counts())
```
